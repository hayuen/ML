### 시각화
* 데이터를 그래프적으로 보여주는 것
* 방대한 양의 자료를 분석하여 한눈에 볼 수 있도록 도표나 차트등으로 정리한 것
### 시각화 차트 5가지
* 순위 / 비교 : bar chart(막대그래프)
* 부분과 전체 : pie chart(도넛 차트)
* 트랜드(추세) : line char(주식시장, 환율변동 등)
* 상관관계 : 산포도(스캐터플롯(x,y좌표를 점으로 찍은것)), 히트맵(열과 행을 색으로 표현)
* 분포 : 히스토그램, box plot(데이터들이 퍼저있는 정도)
>https://matplotlib.org -> tutorials

### figure
* figure : 그래프를 그리기 위한 틀
* axis : figure안에 그려질 틀의 속성을 지정하는 값
* axis가 가지고 있는 요소
    - set_title : 그래프의 제목 부분
    - gird : 틀을 바둑판 처럼 생성
    - legend : 범례
    - x axis label : 가로축 이름
    - y axis label : 세로축 이름
    - tick : 눈금
    - tick label : 눈금에 글자
    - major tick label : 긴 눈금이 들어가는 부분
    - minor tick : 짧은 눈금

### plt.subplots(nrows, ncols)
* plt.subplots를 사용하면 2개의 객체를 돌려준다
    - figure, axis

```python
import warnings
warnings.filterwarnings('ignore')
#애플 사용자 AppleGothic 또는 Aplle Gothic
plt.rc('font', family = 'Malgun Gothic')
```

```python
data_1_x = [1,2,3]
data_1_y = [3,6,9]

data_2_x = [1,2,3]
data_2_y = [2,5,9]

plt.plot(data_1_x, data_1_y)
plt.plot(data_2_x, data_2_y)
plt.title("타이틀 연습")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3) )
ax.plot(data_1_x, data_1_y)
ax.plot(data_2_x, data_2_y)
ax.set_title("타이틀", fontsize=40)

ax.tick_params(axis='y', rotation=90)
ax.tick_params(axis='x', rotation=45)
```

```python

plt.plot(data_1_x, data_1_y)
plt.plot(data_2_x, data_2_y)

plt.xticks( rotation = 90)
plt.yticks( rotation = 45)

plt.title("타이틀 연습")

plt.legend(['범례1, 범례2'], fontsize=20)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3) )
ax.plot(data_1_x, data_1_y, label = "legend-1")
ax.plot(data_2_x, data_2_y, label = "legend-2")
ax.legend( fontsize=20 )
plt.show()
```

#### 스타일 세부 설정 - 마커, 라인, 컬럼
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html 마커의 모양이 있다. 아래로 쭉 내리면 보인다

#### line 설정
* '-' : solid line style
* '--' : dashed line style
* '-.' : dash-dot line style
* ':' : dotted line style
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html 이전 사이트에 들어가면 조금 더 내리면 라인 스타일도 보인다

#### 색상 변경
* b(파란색), g(녹색), r(빨간색), c(청록색), y(노란색), k(검정색), w(흰색)
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html 조금더 내리면 기본 색상도 보인다
* 또는 더 많은 색상으로 보고자 하면 https://matplotlib.org/stable/gallery/color/named_colors.html 색상 이름 똑같이 사용하면 된다

```python
plt.plot(data_1_x, data_1_y, marker='o', markersize=8, linestyle='--', color='b')
plt.plot(data_2_x, data_2_y, marker='x', markersize=8, linestyle=':', color='lime')

plt.xlabel("X축 연습", fontsize=20, color='r')
plt.ylabel("y축 연습")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3))

ax.plot(data_1_x, data_1_y, marker='o', markersize=8, linestyle='--', color='b')
ax.plot(data_2_x, data_2_y, marker='x', markersize=8, linestyle=':', color='lime')

ax.set_xlabel("xxxx", fontsize="20", color="g")
ax.set_ylabel('yyyy')
```

```python
# 1행 2열
fig, ax = plt.subplots(1,2, figsize=(10,3))
ax[0].plot(x,y)
ax[1].scatter(x,y)
plt.show()
```

```python
# 2행 2열
fig, ax = plt.subplots(2,2, figsize=(10,3))
ax[0][0].plot(x,y)
ax[0][1].scatter(x,y)

t_x = [1,2,3,4,5,6,7,8,9]
ax[1][0].hist(t_x, bins = 9)
t_x = [10,20,70]
ax[1][1].pie(t_x, labels=['일','이','삼'])

plt.show()
```
