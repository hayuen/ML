```python
# 다시 보ㄱ기





#모든 결측치가 있는 열에 대해서 해당 대여소 번호에 맞는 평균값으로 치환
#소수점 첫번째 짜리까지 구한후 변경해 주세요 round (x,1)
# 예) 대여거치대가 NaN값들이 존재 한다면 해당 대여거치대의 대여소 번호중 같은 대여소 번호를 가지고 있는 평균을 구해 해당 대여거치대 값을 처리하세요
import numpy as np


nan_list = ['대여거치대','반납거치대','이용시간','이용거리']
for number in df['대여소번호'].unique():
    # number == 646,3,4,56, ..... 이렇게 뜰거란말이지? 그리고 for 문돌려서 name 으로 하나씩 봅는다음에
    # mean = round( df[name][df['대여소번호']==number.fillna(mean) number랑 같으면 mean 평균값으로 바꿔줘라 이거잖아.]  ,1) 뭐 라운드는 소수점자리 1자리로 뽑는다치고
    for name in nan_list:
        mean = round(df[name][df['대여소번호']==number].mean(),1)
        #df[name][df['대여소번호']==number] = df[name][df['대여소번호']==number].fillna(mean)
        #위와 같이 진행하게 되면 복사본을 원본에 설정하는 경고가 뜬다. 그래서 아래처럼 바꾸자
        df.loc[df['대여소번호']==number,name] = df[name][df['대여소번호']==number].fillna(mean)
df.loc[3] #결측값 위치 확인 결과 결측값(반납거치대)이 평균값으로 치환 확인
```
