```python
import pandas as pd
df = pd.read_csv('data/bicycle.csv')
df.head()
```

```python
li = df['대여소번호'].unique() # == 대여소번호 리스트

for i in li:
    mean = round(df['대여거치대'][df['대여소번호']==i].mean(),0)
    df.loc[df['대여소번호']==i,['대여거치대']] = df['대여거치대'][df['대여소번호']==i].fillna(mean)
```

```python
time_dic = df_time.to_dict('list')
time_start_list=time_dic.get('대여소명')
time_end_list=time_dic.get('반납대여소명')
```

```python
#장한평역1번출구에서 중량센터의 값들만 가져온다
# 대여소명과 반납대여소명이 같은 정보들을 가져온다.
import numpy as np
for i in range(df['이용시간'].isnull().sum()) : 
    
    median = df[(df['대여소명'] == time_start_list[i]) & (df['반납대여소명'] == time_end_list[i])]['이용시간'].median()
    df.loc[ (df['대여소명']==time_start_list[i]) & (df['반납대여소명'] == time_end_list[i] ),'이용시간'] = \
        df[(df['대여소명'] == time_start_list[i] ) & (df['반납대여소명'] == time_end_list[i] )]['이용시간'].apply( lambda x : median if np.isnan(x) else x )
```

```python
df_use = df[['대여소번호','반납대여소번호']]
df_use
```

```python
#중복되는 대여소번호와 반납대여소번호 확인
print('연산전 중복데이터 개수 : ',df_use.duplicated().sum())
# df_use.drop_duplicates(inplace=True) 왼쪽의 코드를 사용했더니 경고가 나와서 아래의 내용으로 변경
df_use = df_use.drop_duplicates()
print('연산후 중복데이터 개수 : ',df_use.duplicated().sum())
df_use
```

```python
# 이용시간. 위의 내용을 각각의 리스트 값으로 변환했다.
use_dic = df_use.to_dict('list')
use_start_list=use_dic.get('대여소번호')
use_end_list=use_dic.get('반납대여소번호')
```

```python
def outline( df_re ,rental_num, return_num, name):
    Q1, Q3 = np.percentile(df_re[(df_re['대여소번호']==rental_num) & (df_re['반납대여소번호']==return_num)][name],[25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
for i in range( len(use_start_list) ):
    # 이용시간 중앙값 구하기
    use_start = use_start_list[i]
    use_end = use_end_list[i]
    lower, upper = outline( df ,use_start, use_end, '이용시간')
    print( f"s : {use_start}, e : {use_end} : 이상치 {upper}", end=", 중앙값 : ")
    use_median = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].median()
    print(use_median,"max : ",df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].max())
    df.loc[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end), '이용시간'] = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end)]['이용시간'].apply(lambda x: use_median if x > upper else x )
```
