### 데이터 그룹
* 특정 조건에 맞는 데이터가 하나 이상 존재하는 경우 데이터를 그룹화하여 분석
* 집계함수
    - 그룹 분석에서 사용되는 groupby함수는 그룹을 분류하는 기능을 하며 다양한 그룹 연산 메소드를 지원


```python
import pandas as pd
df = pd.read_csv('data/exam_sample.csv')
df
```

```python
df_class = df.groupby(['class'])
df_class.get_group('A')
```

```python
df_class.mean()
```

```python
df.groupby(['class','sex']).mean()
```

```python
df.groupby(['class']).mean()[['math']]
```

```python
df_class['english'].count()
```

```python
df
```

```python
df_class['english'].sum()
```

```python
import numpy as np
df_class.agg({'math':np.max, 'english':np.min})
```

```python
df_class.agg({'math':np.sum, 'english':np.mean})
```

```python
df_class.agg({'math':[np.max, np.sum], 'english':np.mean})
```

```python
df_class.agg({'math':'sum', 'english':'count'}).reset_index()
```
