### one-hot 인코딩
* 사람이 이해하는 언어를 컴퓨터가 쉽게 이해할 수 있도록 행렬로 표현
* 단어가 존재하는 곳은 1, 아닌곳은 0

```python
import pandas as pd
df = pd.DataFrame({
    'city':['부산','부산','인천','제주','서울'],
    'number':['051','051','032','064','02'],
    'int':[1,2,3,4,5]
})
df
```

```python
df_dummies = pd.get_dummies( df )
df_dummies
```

## Label encoding

```python
df = pd.DataFrame({'fruit':['망고','바나나','사과','앵두']})
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['fruit'])
df['number']= le.transform(df['fruit'])
df


```
